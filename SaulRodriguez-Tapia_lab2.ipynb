The purpose of this dataset (https://archive.ics.uci.edu/dataset/165/concrete+compressive+strength) that I selected
is to predict the strength of concrete

```python
 
#ignore warnings

import warnings

warnings.simplefilter(action = "ignore", category = FutureWarning)
```

```python

#import dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

#import seaborn

import seaborn as sns
```

```python
df = pd.read_excel('Concrete_Data.xls')
#remove leading/trailing spaces from column names
df.columns = df.columns.str.strip()
```

```python
sns.lineplot(
    data=df,
    x='Age (day)',
    y='Concrete compressive strength(MPa, megapascals)',
    ci='sd',   #standard deviation as confidence interval
    marker='o',
    linewidth=2
)
plt.title('Concrete Strength Over Time')
plt.xlabel('Age (days)')
plt.ylabel('Compressive Strength (MPa)')
plt.grid(True)
plt.savefig("SaulRodriguez-Tapia_line.png")
plt.show()
```

This line plot shows the compressive strength over time, the shadow around the line is the confidence interval for the mean. over time the concrete generally gets stronger, except for the spike in strength around the 100th day.

```python
import seaborn as sns
import matplotlib.pyplot as plt

#style to include a grid
sns.set_style("whitegrid")

#scatterplot
sns.scatterplot(
    data=df,
    x='Cement (component 1)(kg in a m^3 mixture)',
    y='Concrete compressive strength(MPa, megapascals)',
    hue='Age (day)',
    palette='viridis'
)

plt.title('Strength vs Cement Content')
plt.xlabel('Cement (kg/m)')
plt.ylabel('Compressive Strength (MPa)')
plt.grid(True)
plt.savefig("SaulRodriguez-Tapia_Scatter.png")
plt.show()
```

This scatterplot shows the compressive strength through increasing cement content at certain days. in general the less cement content and least num of days; the weaker the compressive strength, but more cement content and more aged concrete has a stronger compressive strength

```python

sns.barplot(
    data=df,
    x='Age (day)',
    y='Concrete compressive strength(MPa, megapascals)',
    estimator='mean',
    ci='sd'  # Optional: shows error bars (standard deviation)
)

plt.title('Average Concrete Strength by Age (Days)')
plt.xlabel('Concrete Age (days)')
plt.ylabel('Avg Compressive Strength (MPa)')
plt.xticks(rotation=45)  
plt.tight_layout()
plt.savefig("SaulRodriguez-Tapia_Bar.png")
plt.show()
```

This barplot shows the average compressive strength over certain days and their means

```python
sns.boxplot(
    data=df,
    x='Age (day)',
    y='Concrete compressive strength(MPa, megapascals)'
)

plt.title('Compressive Strength Spread by Age')
plt.xlabel('Age (days)')
plt.ylabel('Compressive Strength (MPa)')
plt.tight_layout()
plt.savefig("SaulRodriguez-Tapia_Boxplot.png")
plt.show()
```

This box plot helps show mean compressive strength over certain days; this shows that concrete at around 90 days old is the strongest on average and with no outliers

```python

components_vs_strength = df[[
    'Cement (component 1)(kg in a m^3 mixture)',
    'Blast Furnace Slag (component 2)(kg in a m^3 mixture)',
    'Fly Ash (component 3)(kg in a m^3 mixture)',
    'Water  (component 4)(kg in a m^3 mixture)',
    'Superplasticizer (component 5)(kg in a m^3 mixture)',
    'Coarse Aggregate  (component 6)(kg in a m^3 mixture)',
    'Fine Aggregate (component 7)(kg in a m^3 mixture)',
    'Concrete compressive strength(MPa, megapascals)'
]]

corr_strength = components_vs_strength.corr()[['Concrete compressive strength(MPa, megapascals)']].drop('Concrete compressive strength(MPa, megapascals)')

plt.figure(figsize=(8, 6))
sns.heatmap(corr_strength, annot=True, cmap='YlGnBu', fmt=".2f", cbar=True)
plt.title('Correlation of Components with Concrete Strength')
plt.xlabel('Correlation with Strength')
plt.tight_layout()
plt.savefig("SaulRodriguez-Tapia_heat.png")
plt.show()
```

this heatmap tracks the correlation of which components in concrete increase or decrease concrete strength, this even shows which components make the concrete weaker or dont even affect it at all.
